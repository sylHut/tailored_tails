```python
# Load the vehicles.csv file into a DataFrame.
file_location = '/FileStore/tables/vehicles.csv'
df = spark.read.csv(file_location, inferSchema=True, header=True)
# Display the DataFrame.
display(df)
```

```python
# Get the number of each model using a groupby. 
count_df = df.groupBy("manufacturer").count()
count_df.show()
```

```python
# Display the DataFrame and create a bar chart.
display(count_df)
```

```python
# Create a temporary view of the DataFrame
df.createOrReplaceTempView("vehicles")
```

```python
%sql
--  Write a query that shows the top five rows of the temporary view.
SELECT *
FROM vehicles
LIMIT 5;
```

```python
%sql
-- Write a query that returns the number of vehicles by manufacturer and create a bar chart.
SELECT manufacturer, COUNT(*) AS vehicle_number
FROM vehicles
GROUP BY manufacturer;
```

```python
# Bonus: Load the parquet file and load the following columns: year, manufacturer, and transmission.
vehicles_parquet_df = spark.read.format("parquet").load("/FileStore/tables/vehicles.parquet").select("year", "manufacturer", "transmission")
 
# Display the data.
display(vehicles_parquet_df)
```

```python
# Get the number of vehicles by manufacturer. 
vehicles_parquet_df2 = vehicles_parquet_df.groupBy("manufacturer").count()

# Display the data and create a bar chart.
display(vehicles_parquet_df2)
```
