```python
# Create variables for the S3 files.
species_data = 's3://2u-data-curriculum-team/dataviz-classroom/v1.2/22-big-data/4/species.csv'
survey_data = 's3://2u-data-curriculum-team/dataviz-classroom/v1.2/22-big-data/4/surveys.csv'
```

```python
# Read in the species.csv file.
species_df = spark.read.csv(species_data, inferSchema=True, header=True)
display(species_df)
```

```python
# Read in the survey.csv file.
survey_df = spark.read.csv(survey_data, inferSchema=True, header=True)
display(survey_df)
```

```python
# Get the number of records in the species_df.
species_df.count()
```

```python
# Get the number of records in the surveys_df.
survey_df.count()
```

```python
# How many birds are there in the dataset?
joined_table = species_df.join(survey_df, species_df.species_id == survey_df.species_id, how='inner')
results = joined_table.filter(species_df.taxa == 'Bird').count()
print(results)
```

```python
# How many rodents were recorded in 1978?
results2 = joined_table.filter(species_df.taxa=='Rodent').filter(survey_df.year==1977).count()
print(results2)
```

```python
# Create a temp view to query in SQL
species = species_df.createOrReplaceTempView('species')
```

```python
# Create a temp view to query in SQL
survey = survey_df.createOrReplaceTempView('survey')
```

```python
%sql
SELECT * 
FROM species 
LIMIT 5;
```

```python
%sql
-- Create a preview of the first five rows.
SELECT *
FROM survey
LIMIT 5;
```

```python
%sql
-- How many birds are in the dataset?
SELECT COUNT(*)
FROM species
INNER JOIN survey
ON species.species_id = survey.species_id
WHERE species.taxa = 'Bird';
```

```python
%sql
-- How many rodents were recorded in 1977?
SELECT COUNT(*)
FROM species
INNER JOIN survey
ON species.species_id = survey.species_id
WHERE species.taxa = 'Rodent'
AND survey.year = 1977;
```
