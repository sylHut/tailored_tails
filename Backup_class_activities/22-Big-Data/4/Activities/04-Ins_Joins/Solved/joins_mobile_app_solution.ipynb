```python
# Load the user_device.csv and user_usage.csv files from an AWS S3 bucket into Spark DataFrames.
user_device = 's3://2u-data-curriculum-team/dataviz-classroom/v1.2/22-big-data/4/user_device.csv'
user_usage = 's3://2u-data-curriculum-team/dataviz-classroom/v1.2/22-big-data/4/user_usage.csv'

user_device_df = spark.read.csv(user_device, inferSchema=True, header=True)
display(user_device_df)

user_usage_df = spark.read.csv(user_usage, inferSchema=True, header=True)
display(user_usage_df)
```

```python
# Merge the two DataFrames on the "use_id" column. 
merged = user_device_df.join(user_usage_df, on=['use_id'], how='inner')

# Display the merged DataFrames
display(merged)
```

```python
# Get the average monthly data usage per platform. 
merged.groupBy("platform").agg({"monthly_mb": "avg"}).show()
```

```python
# Create temporary views of each original DataFrame. 
user_device_df.createOrReplaceTempView('device')
user_usage_df.createOrReplaceTempView('usage')
```

```python
%sql
-- Get the average monthly data usage per platform.
SELECT AVG(monthly_mb), platform
FROM usage
JOIN device
ON usage.use_id = device.use_id
GROUP BY platform;
```
