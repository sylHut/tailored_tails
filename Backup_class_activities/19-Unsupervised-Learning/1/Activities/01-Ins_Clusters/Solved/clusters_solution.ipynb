```python
# Import the modules
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
# Generate three synthetic clusters
X, y = make_blobs(
    centers=3, 
    random_state=1
)
```

```python
# Visualizing all three classes
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
# Generate five synthetic clusters
X, y = make_blobs(
    centers=5, 
    random_state=1
)
```

```python
# Visualizing all five classes
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
# Generate 5 synthetic clusters with a different random state, such as 42
X, y = make_blobs(
    centers=5, 
    random_state=42
)
```

```python
# Visualizing all classes
plt.scatter(X[:, 0], X[:, 1], c=y)
```
